# 必要パッケージのインストール

```python
import pandas as pd
import numpy as np
```

```python
rois_in_image = 60
images = 90
row_num = rois_in_image * images
```

# ot2_ypd/Results.csvからnormalized_ot2-ypd.csvを出力

## ディレクトリ・ファイルの取得

```python
cd /mnt/d/github_repo
```

```python
target_dir = "automatic_spotassay/"
filename = "quantification_with_imageJ/ot2_ypd/Results.csv"
path = target_dir + filename
csv=pd.read_csv(path)
```

```python
df=csv[0:row_num]['Mean'].to_frame()
```

## 加工

```python
df['executer']="ot2"
```

```python
df['agar']="YPD"
```

```python
df['timepoint']=0
```

```python
df[3820:3830]
```

```python
spot_num_in_image = 60 #image一枚につき、どれだけのspotを登録したか
spot_num_in_plate = 30
spot_num_in_serial = 5
```

```python
for i in range(int(len(df)/spot_num_in_image)):
    df['timepoint'][0+i*spot_num_in_image:spot_num_in_image+i*spot_num_in_image]=1+i
```

```python
df['plate']=1
```

```python
for i in range(int(len(df)/spot_num_in_image)):
    df['plate'][0+i*spot_num_in_image:spot_num_in_plate+i*spot_num_in_image]=1
    df['plate'][spot_num_in_plate+i*spot_num_in_image:spot_num_in_image+i*spot_num_in_image]=2
```

```python
df['serial_num']=0
```

```python
for i in range(int(len(df)/spot_num_in_plate)):
    df['serial_num'][0+i*spot_num_in_plate:spot_num_in_serial+i*spot_num_in_plate]=1
    df['serial_num'][spot_num_in_serial+i*spot_num_in_plate:2*spot_num_in_serial+i*spot_num_in_plate]=2
    df['serial_num'][2*spot_num_in_serial+i*spot_num_in_plate:3*spot_num_in_serial+i*spot_num_in_plate]=3
    df['serial_num'][3*spot_num_in_serial+i*spot_num_in_plate:4*spot_num_in_serial+i*spot_num_in_plate]=4
    df['serial_num'][4*spot_num_in_serial+i*spot_num_in_plate:5*spot_num_in_serial+i*spot_num_in_plate]=5
    df['serial_num'][5*spot_num_in_serial+i*spot_num_in_plate:6*spot_num_in_serial+i*spot_num_in_plate]=6
```

```python
df['spot']=0
```

```python
for i in range(int(len(df)/spot_num_in_serial)):
    df['spot'][0+i*spot_num_in_serial]=1
    df['spot'][1+i*spot_num_in_serial]=2
    df['spot'][2+i*spot_num_in_serial]=3
    df['spot'][3+i*spot_num_in_serial]=4
    df['spot'][4+i*spot_num_in_serial]=5
```

```python
df['normalized']=float(0)
```

```python
for i in range(int(len(df))):
  df['normalized'][i] = df['Mean'][i]/df['Mean'][i%60]
```

```python
print(df.info())
```

```python
df
```

##   csvに出力

```python
target_dir
```

```python
df.to_csv(target_dir + 'data_analysis/ot2_ypd_normalized.csv')
```
